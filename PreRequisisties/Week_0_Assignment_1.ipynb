### Question 2
 Create a list of 10 random numbers between 1-100. Create a balanced BST (Binary search tree) from this array

````python
from typing import List
from dataclasses import dataclass
import random
import rich

@dataclass
class Node:
    value: int
    left: None
    right: None


class Tree:
    def __init__(self, input_array: List[int]) -> None:
        self.root = None
        self._build_tree(input_array)

    def _build_tree(self, values_array: List[int]):
        for value in values_array:
            self.root = self.insert_node(self.root,value)

    def insert_node(self, root: Node, value: int):
        if root is None:
            root = Node(value, left=None, right=None)
        else:
            if root.value == value:
                return root
            elif root.value < value:
                root.right = self.insert_node(root.right, value)
            else:
                root.left = self.insert_node(root.left, value)
        return root

    def inorder(self,root):
        if root:
            self.inorder(root.left)
            print(root.value,end=" ")
            self.inorder(root.right)
    def inorder_wrapper(self,root):
        print(f"inorder traveral of tree starting from node {root.value} ⮯")
        self.inorder(root)


random_array = [random.randint(0,100) for i in range(random.randint(0,100))]
t = Tree(random_array)
t.inorder_wrapper(t.root)
t.insert_node(t.root,21)
t.inorder_wrapper(t.root)
````

### Question 3
Given the following dictionary:


```python
{"black": "r", "hero": "e", "go": "g", "clue": "i", "mean": "q", "groan": "o",
             "sin": "p", "pint": "u", "tone": "n", "graze": "s", "sea": "t", "plant": "a"}
```

Create a list of all the keys of the dictionary, which can be formed from the values present in the dictionary.

````python
test_dict = {"black": "r", "hero": "e", "go": "g", "clue": "i", "mean": "q", "groan": "o",
             "sin": "p", "pint": "u", "tone": "n", "graze": "s", "sea": "t", "plant": "a"}
formed_words_list = []
values = list(test_dict.values())

for key in test_dict.keys():
    letters_key = list(key)
    check = all(item in values for item in letters_key)
    if check:
        formed_words_list.append(key)

print(formed_words_list)
````

Given the following list of tuples,


```python
listA = [(1,2), (4,3), (2,10), (12, 5), (6, 7), (9,11), (15, 4)]
```
* sort the list by 1st item in the tuple in ascending order

* Sort the list by 2nd item in the tuple in ascending order

* Repeat a, b using descending order

````python
from operator import itemgetter

lista = [(1, 2), (4, 3), (2, 10), (12, 5), (6, 7), (9, 11), (15, 4)]

# sort the list by 1st item in the tuple in ascending order
print(f"sort the list by 1st item in the tuple in ascending order ➡️ ${sorted(lista,key=itemgetter(0))}")

# Sort the list by 2nd item in the tuple in ascending order
print(f"Sort the list by 2nd item in the tuple in ascending order ➡️ ${sorted(lista,key=itemgetter(1))}")

# sort the list by 1st item in the tuple in descending order
print(f"sort the list by 1st item in the tuple in descending order ➡️ ${sorted(lista,key=itemgetter(0),reverse=True)}")

# Sort the list by 2nd item in the tuple in descending order
print(f"Sort the list by 2nd item in the tuple in descending order ➡️ ${sorted(lista,key=itemgetter(1),reverse=True)}")
````

### Question 5
What do you mean by a mutable and immutable object. Are string, list, tuple, dictionary mutable type objects or immutable objects.

Mutable object modify the object inplace ,i.e.  they modify the calling object directly in memory

Example ➜ Calling ```.reverse``` on a list changes the order of the underlying list 
```python
>>> li = [1, 2, 3, 4]
>>> li.reverse() 
>>> li
[4, 3, 2, 1] 
```

Immutable objects, by contrast, do not change through a function. Calling a function on the object does not change the object instead returns a new value

Example ➜ Calling ```.lower``` on a string returns a new string with lowercase letters
```python
>>> langauge = "JavaScript"
>>> language.lower()
"javascript"
>>> langauge
"JavaScript"
```

| Data type  | Mutable |
|------------|---------|
| string     | No      |
| list       | yes     |
| tuple      | No      |
| dictionary | Yes     |
